```python
from data_prep import DataPrep
from return_forecasting import ReturnForecast

from dimension_redux import DimensionRedux

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay,mean_squared_error
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVR
from sklearn.linear_model import LinearRegression
import xgboost as xgb

import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, confusion_matrix, ConfusionMatrixDisplay

import pandas as pd
import numpy as np
```

```python
dr_obj = ReturnForecast()
dr_obj.data_initialization()
dr_obj.data_construction()

pca_reduced_df = dr_obj.pca_fit()
```

```python
# PCA Random Forest

sectors = list(pca_reduced_df['sector'].unique()) + ['all']
# Use a colormap or define a custom color palette
colors = plt.cm.viridis(np.linspace(0, 1, len(sectors)))

fig, ax = plt.subplots(figsize=(12, 8))

performance_metrics_dict = {}

x_labels = [i for i in pca_reduced_df.columns if i not in (dr_obj.id_columns + dr_obj.ret_col + ['ret_class', 'index'])]
for sector, color in zip(sectors, colors):
    if sector == 'all':
        sub_stock_data = pca_reduced_df
    else:
        sub_stock_data = pca_reduced_df.loc[pca_reduced_df.sector == sector]
    sub_stock_df_x = sub_stock_data[x_labels]
    sub_stock_df_y_dis = sub_stock_data['ret_class']
    X_test, y_test, predictions = dr_obj.random_forest_fit(x_data=sub_stock_df_x, y_data=sub_stock_df_y_dis, title_suffix=f'{sector}+pca+random_forest')
    signal_df, predict_df = dr_obj.daily_level_portfolio_return_construction(predictor_type='discrete', y_test=y_test, predictions=predictions, fee_rate=0.001)
    
    performance_metrics_dict[sector] = {}
    performance_metrics_dict[sector]['std'] = signal_df['signal_ret'].std() * np.sqrt(12)
    years = (predict_df.index[-1] - predict_df.index[0]).days / 365.25
    performance_metrics_dict[sector]['ret'] = float(predict_df.values[-1][0] - 1) / years
    performance_metrics_dict[sector]['annualized_sharpe_ratio'] = performance_metrics_dict[sector]['ret'] / performance_metrics_dict[sector]['std']

    ax.plot(predict_df, label=sector, color=color)

ax.legend(loc='upper left')
ax.set_title('PCA Simulated Returns - Random Forest')
print(pd.DataFrame(performance_metrics_dict).T)
fig
```
